```
import numpy as np
```

```
mu, sigma = 0, 0.1 # mean and standard deviation
s = np.random.normal(mu, sigma, 1000)
```

```
print(s.shape)
```

```
np.mean(s)
```

```
np.var(s)
```

```
import matplotlib.pyplot as plt
plt.plot(s)
```

```
np.histogram(s)
```

```
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(s, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```
import scipy
from scipy import stats
plt.plot(scipy.stats.norm.pdf(s), color='r')
plt.plot(scipy.stats.norm.cdf(s), color='b')
```

```
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(scipy.stats.norm.pdf(s), 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```
import matplotlib.pyplot as plt
count, bins, ignored = plt.hist(scipy.stats.norm.cdf(s), 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
```

```
print((s>0.001).mean())
```
